## Downloading files

<p> Para comenzar con la limpieza de los datos es imprescindible saber desde donde se limpiarán y almacenarán los datos.<br></br> 
Las funciones **"getwd()"** y **"setwd()"** resultan útiles para setear el directorio de trabajo, gracias a getwd() podemos movernos y ubicar en que dirección actualmente se encuentra el usuario, mientras que con setwd() fijamos el directorio de trabajo.</p>

getwd() = retorna la url en la que se encuentra el usuario  <br>
setwd() = fija la url del usuario

Para no trabajar con url tan grandes se puede trabajar con direcciones relativas:
setwd("./data") o setwd("../") con esto nos movemos arriba un directorio

```r
# Ejemplo descargar data, importar y crear nuevo directorio
url_download <- "http://data.baltimorecity.gov/api/views/htv2-xgaj/rows.csv?accessType=DOWNLOAD&bom=true&format=true&delimiter=%3B"
if (!file.exists("./data")){
    dir.create("./data")
}
download.file(url = url_download, destfile = "./data/agua.csv")
# parametro method = curl es para MAC
# destfile => es el dir donde se importará la data. 
list.files("./data")
```

```r
dateDownloaded <- date()
dateDownloaded 
# lo mas probable es que los datos descargados puedan cambiar, actualizarse en el futuro, por lo que es
# es recomendable guardar la fecha de la descargar, para tener seguimiento de cualquier actualización de la data.
```

## Archivos XML 

Es la base del web scrapping.
Frecuentemente usada para guardar data estructurada 
Es un Markup language
Componentes: 
- markup = etiqueta que dan la estructura al texto 
- content = el contenido, texto. 
<section> </section>
<line-break />
<Greeting> </Greeting> => etiquetas especificas 
<img src=" " alt=" " /> 
<step number = "3"> </step> 

```r
install.packages("XML", dependencies = TRUE)
```

```r
library(XML)
xml_url <- "https://data.baltimorecity.gov/api/views/k5ry-ef3g/rows.xml?accessType=DOWNLOAD"
if(!file.exists("./data/restoranes.xml")){
    download.file(xml_url,destfile = "./data/restoranes.xml")
}
doc <- xmlTreeParse(file = "./data/restoranes.xml",useInternal= TRUE)
rootNode <- xmlRoot(doc)
xmlName(rootNode)
```

```r
names(rootNode)
```

```r
rootNode[[1]]
```

```r
rootNode[[1]][[1]][[2]]
```

```r
rootNode[[1]][[1]]
```

```r
doc
```

```r
# programatically extracts part of a file 
xmlSApply(rootNode,xmlValue) # itera por todos los elementos del rootNode y retorna el valor de la funcion xvalue
```

```r
# xpath otra forma de recorrer archivos XML.
xpathSApply(rootNode,"//neighborhood",xmlValue) # extraemos el contenido de las etiquetas neighborhood
```

```r
library(XML)
library(httr)
web_url <- "https://www.portalinmobiliario.com/arriendo/casa/concon-valparaiso"
response = GET(web_url) # install.packages("httr")
data_arriendos <- htmlTreeParse(response,useInternal = TRUE)
```

```r
rootNode <- xmlRoot(data_arriendos)  # para empezar a recorrer el archivo, reconoce la etiqueta inicial 
names(rootNode[[1]]) # entrega el nombre de las estiquetas dentro de la posicion 1 => header 
names(rootNode[[2]]) # entrega las etiquetas dentro del <body> 
```

```r
response <- GET("https://www.espn.com/nfl/team/_/name/bal/baltimore-ravens")
doc <- htmlTreeParse(response,useInternal=TRUE)
```

```r
jugadores <- xpathSApply(doc,"//a[@name='&lpos=nfl:teamclubhouse:draftpicks:player']",xmlValue)
jugadores
# con //a buscamos todas las etiquetas <a> luego [@ especificamos el atributo]
```
